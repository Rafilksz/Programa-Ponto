```

import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo não direcionado
G = nx.DiGraph()

# Definindo as conexões entre os nós
G.add_edge('A', 'E')
G.add_edge('A', 'C')
G.add_edge('B', 'D')
G.add_edge('B', 'A')
G.add_edge('C', 'B')
G.add_edge('C', 'F')
G.add_edge('D', 'F')
G.add_edge('E', 'F')

plt.figure(figsize=(8,6))

nx.draw(G, with_labels=True, node_size=700, node_color='skyblue', font_size=20, font_weight='bold',arrowstyle='->',arrowsize=20)
plt.show()



```

```

def verifica_caminho(G, caminho):
  return all(G.has_edge(caminho[i], caminho[i+1]) for i in range(len(caminho) -1))

caminho = ['A', 'C', 'F']

existe_caminho = verifica_caminho(G,caminho)

print(f"Caminho {caminho} existe" if existe_caminho else f"Caminho {caminho} não existe")
```
