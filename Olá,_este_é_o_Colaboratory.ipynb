```

import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo não direcionado
G = nx.Graph()

# Definindo as conexões entre os nós
connections = [('A', 'B'), ('A', 'C'), ('A', 'F'),
               ('B', 'D'), ('B', 'A'),
               ('C', 'A'), ('C', 'D'), ('C', 'F'),
               ('D', 'E'), ('D', 'C'), ('D', 'B'),
               ('E', 'D'), ('E', 'F'),
               ('F', 'A'), ('F', 'C'), ('F', 'E')]

# Adicionando as conexões ao grafo
G.add_edges_from(connections)

# Plotando o grafo
nx.draw(G, with_labels=True, node_color='lightblue', node_size=1000, font_size=12, font_weight='bold', edge_color='gray')
plt.show()
```
